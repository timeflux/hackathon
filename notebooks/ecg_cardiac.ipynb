# ECG - CARDIAC: read and plot the data from HDF5

```python
# import useful packages
import os

import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# read hdf5 
fname = '../data/bitalino_ecg_A1.hdf5'
with pd.HDFStore(fname, 'r') as store: 
    print(store.keys())
    bitalino = pd.read_hdf(store, '/bitalino')
    cardiac_rr = pd.read_hdf(store, '/cardiac/rr')
    cardiac_features = pd.read_hdf(store, '/cardiac/features')
```

```python
f, ax = plt.subplots(3, 1, sharex=True, figsize=(20, 15))
ax[0].plot(bitalino.A1)
ax[0].legend(['raw ecg'], loc=1)
ax[1].plot(60/cardiac_rr)
ax[1].legend(['bpm'], loc=0)
ax[2].plot(cardiac_features); 
ax[2].legend(cardiac_features.columns,  loc=0)
plt.suptitle(f'Data from file {os.path.basename(fname)}')
```
